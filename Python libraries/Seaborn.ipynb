# Line plot

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.simplefilter("ignore")
```

```python
var=[1,2,3,4,5]
var_1=[3,5,4,6,7]
df=pd.DataFrame({"var":var,"var_1":var_1})
df
```

```python
plt.figure(figsize=(5,4))
plt.plot(var,var_1)
plt.show()
```

```python
# sns works with dataframe
# sns.linepltot(x="var",y="var_1")
```

```python
plt.figure(figsize=(5,4))
sns.lineplot(x="var",y="var_1",data=df)
plt.show()
```

```python
y1=sns.load_dataset("penguins")
y1.head(10)
```

```python
y1.dropna(inplace= True)
```

```python
y1.isnull().sum()
```

```python
sns.lineplot(x="bill_depth_mm",y="flipper_length_mm",data=y1,hue="sex")
plt.show()
```

# histogram

```python
sns.displot(x="bill_depth_mm",data=y1,bins=[14,16,18,20,22],kde=True,rug=True,color="g") #log_scale=True
plt.show()
```

# Bar plot

```python
sns.barplot(x="island",y="bill_depth_mm",data=y1,hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=y1,hue="sex",ci=10) #oriention requires both x and y numeric value
plt.show()
```

```python
sns.barplot(x="flipper_length_mm",y="bill_depth_mm",data=y1,orient="h")
plt.show()
```

```python
sns.set(style="dark")
sns.barplot(x="island",y="bill_depth_mm",data=y1,palette="crest",saturation=50,errcolor="y",alpha=0.5)
plt.show()
```

# scatter plot

```python
var=sns.load_dataset("penguins").head(40)
```

```python
# sns.set(style="darkgrid")
# sns.set(style="dark")
m={"Male":"*","Female":"o"}
sns.scatterplot(x="flipper_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(100,60),palette="gist_ncar",markers=m)
plt.show()
```

# Heat Map

```python
var1=sns.load_dataset("anagrams")
```

```python
var1.drop(columns="attnr",axis=1,inplace=True)
var=var1.head(10)
var
```

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var,vmin=0,vmax=12,cmap="crest",annot=True,annot_kws=y)
plt.show()
```

```python
var2=np.linspace(1,10,10).reshape(2,5)
var2
```

```python
ar=np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])
v=sns.heatmap(var2,vmin=0,vmax=10,cmap="crest",annot =ar,fmt="s",linewidth=10,linecolor="r",cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="python",ylabel="wscube")
# sns.set(font_scale=5)
plt.show()
```

# Pairplot

```python
var3=sns.load_dataset("tips")
var3.head()
```

```python
sns.set(style="dark")
sns.pairplot(var3,hue="sex")
# sns.set(font_scale=1)
plt.show()
```

```python
sns.pairplot(var3,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="gist_ncar")
plt.show()
```

```python
sns.pairplot(var3,hue="sex",kind="reg")  # kind can kde, hist,scatter, reg
plt.show()
```

```python
sns.pairplot(var3,hue="sex",markers=["*",">"])
plt.show()
```
