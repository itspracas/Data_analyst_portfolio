# Group by and Aggregating 

```python
df7=pd.read_csv('Flavors.csv')
df7.head()
```

```python
group_by_frame=df7.groupby('Base Flavor').mean(numeric_only=True)
```

```python
group_by_frame.mean()
```

```python
df7.groupby('Base Flavor').count()
```

```python
df7.groupby('Base Flavor').min()
```

```python
df7.groupby('Base Flavor').max()
```

```python
df7.groupby('Base Flavor').sum(numeric_only=True)
```

```python
df7.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
# grouping multiple columns
df7.groupby(['Base Flavor','Liked']).mean(numeric_only=True)
```

```python
df7.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
df7.groupby('Base Flavor').describe()
```

# Merge, Join and Concatenate 

```python
df8=pd.read_csv('LOTR.csv')
df8
```

```python
df9=pd.read_csv('LOTR 2.csv')
df9
```

```python
df8.merge(df9)  # default inner join
```

```python
df8.merge(df9,on = ['FellowshipID','FirstName'],how='inner')    
```

```python
df8.merge(df9,how='outer')
```

```python
df8.merge(df9,how='left')
```

```python
df8.merge(df9,how='right')
```

```python
df8.merge(df9,how='cross')
```

```python
df8.join(df9,on='FellowshipID',how='outer',lsuffix='_left',rsuffix='_right')
```

```python
pd.concat([df8,df9])
```

```python
pd.concat([df8,df9],join='outer',axis=1)
```
